```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#loading the random datasets with  6 features among with 3 are informative
x,y = make_regression(n_samples=100,n_features=6,n_informative=2,random_state=0,shuffle=False)
```

```python
#finding value in x
x
```

```python
#finding value in  y
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.15)
```

```python
#creating an object of random forest regressor
regr = RandomForestRegressor(max_depth=2,random_state=0)
```

```python
#training the model
regr.fit(x_train,y_train)
```

```python
pred_reg=regr.predict(x_test)
```

```python
r2_score(y_test,pred_reg)
```
